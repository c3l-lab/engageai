```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import os
from dotenv import load_dotenv

load_dotenv()
# # === Credentials ===
EMAIL = os.getenv("EMAIL")
PASSWORD = os.getenv("PASSWORD")

# === Setup Chrome ===
options = Options()
options.add_experimental_option("detach", True)  # Keeps browser open after script ends
driver = webdriver.Chrome(options=options)
driver.maximize_window()

# === Navigate to Azure Synapse ===
driver.get("https://web.azuresynapse.net/en/home?workspace=%2Fsubscriptions%2F8a540307-07bc-46a8-890c-82d4174e323a%2FresourceGroups%2Fprd-workspace-rg%2Fproviders%2FMicrosoft.Synapse%2Fworkspaces%2Fprd-analyticsworkspace")

try:
    wait = WebDriverWait(driver, 10)

    # Step 1: Enter email
    email_input = wait.until(EC.presence_of_element_located((By.NAME, "loginfmt")))
    email_input.send_keys(EMAIL)

    next_button = wait.until(EC.element_to_be_clickable((By.ID, "idSIButton9")))
    next_button.click()
    print("✅ Email entered.")

    # Step 2: Enter password
    password_input = wait.until(EC.presence_of_element_located((By.NAME, "Password")))
    password_input.send_keys(PASSWORD)
    signin_button = wait.until(EC.element_to_be_clickable((By.ID, "submitButton")))
    signin_button.click()
    print("✅ Password entered and Sign in clicked.")

    # # Step 3: Handle "Stay signed in?"
    # try:
    #     stay_signed_in_button = wait.until(EC.element_to_be_clickable((By.ID, "idSIButton9")))
    #     stay_signed_in_button.click()
    #     print("✅ Stay signed in - clicked.")
    # except:
    #     print("ℹ️ Stay signed in screen not shown.")

    try:
        # Click "Don't show this again" checkbox
        dont_show_checkbox = wait.until(EC.element_to_be_clickable((By.ID, "KmsiCheckboxField")))
        if not dont_show_checkbox.is_selected():
            dont_show_checkbox.click()
        print("✅ 'Don't show this again' checkbox selected.")
        
        # Click "Yes" to stay signed in
        stay_signed_in_button = wait.until(EC.element_to_be_clickable((By.ID, "idSIButton9")))
        stay_signed_in_button.click()
        print("✅ Clicked 'Yes' to stay signed in.")
    except Exception as e:
        print(f"❌ Error during 'Stay signed in' step: {e}")

    # # Step 4: Optional - Handle 2FA push (if exists)
    # try:
    #     push_button = wait.until(EC.element_to_be_clickable((By.XPATH, "//input[@value='Send push' and contains(@class, 'Send Push')]")))
    #     push_button.click()
    #     print("✅ 2FA push sent.")
    # except:
    #     print("ℹ️ 2FA push not required or already handled.")

    # === Step 5: Handle "big screen with button in middle" (e.g. Launch Synapse Studio) ===
    try:
        launch_button = wait.until(EC.element_to_be_clickable((By.XPATH, "//button[contains(text(), 'Launch')]")))
        launch_button.click()
        print("✅ Clicked main launch button.")
    except:
        print("ℹ️ No main launch button found — may already be in Synapse.")

    # === Step 6: Handle GitHub login popup (click "Cancel") ===
    # try:
    #     github_popup_cancel = wait.until(EC.element_to_be_clickable((By.NAME, "Cancel")))
    #     github_popup_cancel.click()
    #     print("✅ GitHub login popup canceled.")
    # except:
    #     print("ℹ️ GitHub popup not shown or already closed.")

    try:
        main_window = driver.current_window_handle
        wait.until(lambda d: len(d.window_handles) > 1)
        for handle in driver.window_handles:
            if handle != main_window:
                driver.switch_to.window(handle)
                break

        # Then click Cancel
        github_popup_cancel = wait.until(EC.element_to_be_clickable((By.NAME, "Cancel")))
        github_popup_cancel.click()

        # Switch back
        driver.switch_to.window(main_window)
    except: 
        print("No cancal being clicked")


    # === Step 7: Click "Develop" tab ===
    develop_button = wait.until(EC.element_to_be_clickable((By.XPATH, "//div[contains(text(), 'Develop') and contains(@class, 'expanded-button-label')]")))
    develop_button.click()
    print("✅ Clicked 'Develop' tab.")

    # === Step 8: Click on the 'Engage_AI_SQL' folder ===
    engage_ai_sql = wait.until(EC.element_to_be_clickable((By.XPATH, "//span[contains(text(), 'Engage_AI_SQL')]")))
    engage_ai_sql.click()
    print("✅ Clicked 'Engage_AI_SQL' folder.")

except Exception as e:
    print("❌ Error during automation:", e)

# === Keep browser open ===
input("✅ Script finished. Press Enter to close the browser manually...")
```
